```python
import pandas 
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
from sklearn.linear_model import LinearRegression
from numpy import linspace 
%matplotlib inline

smap = { # in-well substrate concentrations in micromolar 
    'A': 300, 
    'B': 150, 
    'C': 75, 
    'D': 37.5, 
    'E': 18.75, 
    'F': 9.375, 
    'G': 4.6875,
    'H': 0, 
}
```

```python
df = pandas.read_csv( 'titration_column_3.csv' )

df['s'] = df.well.str[0].map( smap )
df['posmaxv'] = - df.maxv
plt.scatter( df.s, df.posmaxv, alpha=0.5, color='r' )
plt.title( '2.7 nM adenosine deaminase' )
```

```python
trip = pandas.read_csv( 'triplicate_1000x_dilution.csv' )
trip['s'] = trip.well.str[0].map( smap )
trip['posmaxv'] = - trip.maxv

def v( s, kcat, km ): return ( kcat * s ) / ( km + s )
popt, pcov = curve_fit( v, trip.s, trip.posmaxv, p0=(50,50) )

x = range( 350 ) 
y = [ v( i, popt[0], popt[1] ) for i in x ]

plt.plot( x, y, color='g' )
plt.scatter( x=trip.s, y=trip.posmaxv, color='k', alpha=0.5 )
plt.title( 'Adenosine deaminase (2.7 nM)' )
plt.xlabel( 'Adenosine (uM)' )
plt.ylabel( 'Rate observed (mOD/min)' )
plt.text( 200, 4, '$k_{cat}$: %0.0f mOD/min' % ( popt[0] ) )
plt.text( 200, 11, '$K_M$: {:0.0f} uM'.format( popt[1] ) )
```

```python
# standard curve
standard_curve = pandas.read_csv( 'standard_curve_edited.csv' )
X = standard_curve.a265[:20] 
y = standard_curve.adenosine[:20]

reg = LinearRegression()
reg.fit( X.reshape( len(X), 1 ), y )

preds = [ reg.predict( i ) for i in [0, 1, 2, 3] ]

plt.scatter( X, y, alpha=0.5, color='k' )
plt.plot( [0, 1, 2, 3], preds, color='r' )
plt.title( 'Adenosine-inosine standard curve' )
plt.xlabel( 'mOD' )
plt.ylabel( 'Adenosine (uM)' )
plt.ylim( 0, 320 )
plt.twinx()
plt.ylabel( 'Inosine (uM)', rotation=-90 )
plt.xlim( 0, 3 )

plt.yticks( linspace( 0, standard_curve.inosine[:20].max(), 7 ).astype( int ) )
plt.gca().invert_yaxis()

print '[adenosine] uM = {:0.0f} * mOD - {:0.0f}'.format( reg.coef_[0], -reg.intercept_ )
```

```python
trip = pandas.read_csv( 'triplicate_1000x_dilution.csv' )
trip['s'] = trip.well.str[0].map( smap ) * 10 **(-9) # converts to mol/L
enzyme_conc = 0.1 / 46000 / 1000 # mol/L = M (in well) 
trip['k'] = ( - trip.maxv ) * 130 * 10**-6 / enzyme_conc # (mOD/min)(uM/mOD)(M/uM)/(M) = 1/min

def v( s, kcat, km ): return ( kcat * s ) / ( km + s )
popt, pcov = curve_fit( v, trip.s, trip.k, p0=(10**7,0.0001) )

from numpy import linspace
x = linspace( 0, trip.s.max(), 100 )
y = [ v( i, popt[0], popt[1] ) for i in x ]

plt.plot( x, y, color='g' )
plt.scatter( x=trip.s, y=trip.k, color='k', alpha=0.5 )
plt.title( 'Adenosine deaminase (enzyme = 2.7 nM)' )
plt.xlabel( 'Adenosine (M)' )
plt.ylabel( 'Rate observed (min$^{-1}$)' )
plt.text( 0.00000015, 500000, '$k_{cat}$: %0.0f $x 10^{6}$ M$^{-1}$ min$^{-1}$' % ( popt[0] * 10**-6 ) )
plt.text( 0.00000015, 100000, '$K_M$: {:0.3f} uM'.format( popt[1] * 10**6 ) )
plt.show()
```
